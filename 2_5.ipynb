```python
# Напишите программу, которая принимает числа и выводит наибольшее число. На вход подаются значения через пробел.


Stroka1 = input() # ввели строку 1
Spisok1 = Stroka1.split() # преобразовали в список 1
Spisok1 = list(map(int, Spisok1))
Maksimum = max(Spisok1)
print(Maksimum)
```
