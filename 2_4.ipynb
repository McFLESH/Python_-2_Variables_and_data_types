```python
# Найти общие и уникальные элементы двух списков продуктов, представленных в виде множеств.
# На вход подаются две строки — продукты первого списка и продукты второго списка через пробел.
#Нужно вывести две строки: общие элементы, а затем уникальные через пробел.
# Общие и уникальные элементы должны быть отсортированы в алфавитном порядке.

Stroka1 = input() # ввели строку 1
Stroka2 = input() # ввели строку 2
Spisok1 = Stroka1.split() # преобразовали в список 1
Spisok2 = Stroka2.split() # преобразовали в список 2
All = Spisok1 + Spisok2 # объединили
All = set(All) # все без повторений

Final2 = list(All.difference(Spisok1)) + list(All.difference(Spisok2))  # оставили уникальные
Final1 = list(All.difference(Final2)) # оставили общие

Final1.sort() # отсортировали

Final2.sort() # отсортировали


Final1 = ' '.join(Final1) # преобразовали в строку

Final2 = ' '.join(Final2) # преобразовали в строку

print(Final1)
print(Final2)
```
