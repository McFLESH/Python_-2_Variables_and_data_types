```python
# Спросить у пользователя, какое количество единиц товара он хочет приобрести и их стоимость, вывести итоговую стоимость покупки.

Number = int(input('введите количество единиц товара: '))
Price = int(input('введите цену товара: '))
Final = Number * Price
print(f'Итого: {Final} руб.')
```
