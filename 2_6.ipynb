```python
# Предложить пользователю подписаться на рассылку новостей о продукте и вывести сообщение о том, что его подписка была зарегистрирована.
# Если вход пользователя неразборчив, то ответить "Не понял Вас, попробуйте ещё раз."
# Для решения данной задачи необходимо использовать словарь:

answers = {
    'Да': 'Вы успешно подписались на рассылку новостей о продукте.',
    'Нет': 'Вы отказались от подписки на рассылку новостей о продукте.'
}
Option = "Не понял Вас, попробуйте ещё раз."

Question = input('Желаете подписаться на рассылку новостей о продукте?')
print(answers.get(Question, Option))
```
